```python
import pandas as pd
```

### Data Collection

```python
data = pd.read_csv("uber_data.csv")
```

```python
data.head()
```

### Data Cleaning
**1.Check if there are any null values <br>
2.Check the datatypes of all columns**

```python
data.info()
```

**Change datatype of pickupdatetime and dropoffdatetime column from object to datetime**

```python
data['tpep_pickup_datetime'] = pd.to_datetime(data['tpep_pickup_datetime'])
data['tpep_dropoff_datetime'] = pd.to_datetime(data['tpep_dropoff_datetime'])
```

```python
data.info()
```

**Drop if any duplicated data is present

```python
data.drop_duplicates().reset_index(drop=True)
```

### Data Manipulation

**1. Create Star Schema**

**Create Tables and necessary columns according to Schema above.**

```python
data['trip_id'] = data.index.astype(str)
```

### Create Dimension Tables
**1. Datetime**

```python
datetime = data[['tpep_pickup_datetime', 'tpep_dropoff_datetime']].reset_index(drop=True)
datetime
```

```python
datetime['tpep_pickup_datetime'] = data['tpep_pickup_datetime']
datetime['tpep_dropoff_datetime'] = data['tpep_dropoff_datetime']

datetime['pickup_hour'] = data['tpep_pickup_datetime'].dt.hour
datetime['pickup_day'] = data['tpep_pickup_datetime'].dt.day
datetime['pickup_weekday'] = data['tpep_pickup_datetime'].dt.weekday
datetime['pickup_month'] = data['tpep_pickup_datetime'].dt.month
datetime['pickup_year'] = data['tpep_pickup_datetime'].dt.year

datetime['dropoff_hour'] = data['tpep_dropoff_datetime'].dt.hour
datetime['dropoff_day'] = data['tpep_dropoff_datetime'].dt.day
datetime['dropoff_weekday'] = data['tpep_dropoff_datetime'].dt.weekday
datetime['dropoff_month'] = data['tpep_dropoff_datetime'].dt.month
datetime['dropoff_year'] = data['tpep_dropoff_datetime'].dt.year
```

```python
datetime['datetime_id'] = datetime.index.astype(str)
datetime = datetime[['datetime_id',
                    'tpep_pickup_datetime',
                    'tpep_dropoff_datetime',
                    'pickup_hour',
                    'pickup_day',
                    'pickup_weekday',
                    'pickup_month',
                    'pickup_year',
                    'dropoff_hour',
                    'dropoff_day',
                    'dropoff_weekday',
                    'dropoff_month',
                    'dropoff_year',
                    ]]

datetime
```

**2. Pickup_location**

```python
pickup_location = data[[ 'pickup_longitude', 'pickup_latitude']].reset_index(drop=True)

pickup_location['pickup_location_id'] = 'P' + (pickup_location.index + 1).astype(str).str.zfill(6)

pickup_location = pickup_location[['pickup_location_id','pickup_longitude', 'pickup_latitude']]
pickup_location
```

**3. Dropoff Location** 

```python
dropoff_location = data[['dropoff_longitude', 'dropoff_latitude']].reset_index(drop=True)

dropoff_location['dropoff_location_id'] = 'D' + (dropoff_location.index + 1).astype(str).str.zfill(6)

dropoff_location = dropoff_location[['dropoff_location_id','dropoff_longitude', 'dropoff_latitude']]
dropoff_location
```

**4. Trip Distance**

```python
trip_distance = data[['trip_distance']].reset_index(drop=True)
trip_distance['trip_distance_id'] = range(100001, 100001 + len(trip_distance))
trip_distance['trip_distance_id'] = trip_distance['trip_distance_id'].astype(str)
trip_distance = trip_distance[['trip_distance_id', 'trip_distance']]

trip_distance
```

**5. Passenger count** 

```python
passenger_count = data[['passenger_count']].reset_index(drop=True)
passenger_count['passenger_count_id'] = 'C' +  (passenger_count.index + 1).astype(str).str.zfill(6)

passenger_count = passenger_count[['passenger_count_id','passenger_count']]

passenger_count
```

**6. Payment Type**

```python
payment_name = {
    1:"Credit card",
    2:"Cash",
    3:"No charge",
    4:"Dispute",
    5:"Unknown",
    6:"Voided trip"
}

payment_type = data[['payment_type']].reset_index(drop=True)
payment_type['payment_type_id'] = 'Pay' +  (payment_type.index + 1).astype(str).str.zfill(6)
payment_type['payment_name'] = payment_type['payment_type'].map(payment_name)

payment_type = payment_type[['payment_type_id','payment_type', 'payment_name']]

payment_type
```

**7. Rate Code Type**

```python
rate_code_type = {
    1:"Standard rate",
    2:"JFK",
    3:"Newark",
    4:"Nassau or Westchester",
    5:"Negotiated fare",
    6:"Group ride"
}

rate_code = data[['RatecodeID']].reset_index(drop=True)
rate_code['rate_code_id'] = rate_code.index.astype(str)
rate_code['rate_code_name'] = rate_code['RatecodeID'].map(rate_code_type)
rate_code = rate_code[['rate_code_id','RatecodeID','rate_code_name']]

rate_code
```

### Create Fact Table

```python
fact_table = pd.DataFrame({
    'trip_id': data['trip_id'],
    'datetime_id': datetime['datetime_id'],
    'pickup_location_id': pickup_location['pickup_location_id'],
    'passenger_count_id': passenger_count['passenger_count_id'],
    'dropoff_location_id': dropoff_location['dropoff_location_id'],
    'trip_distance_id': trip_distance['trip_distance_id'],
    'rate_code_id': rate_code['rate_code_id'],
    'payment_type_id': payment_type['payment_type_id'],
})
```

```python
fact_table = fact_table.merge(datetime, on='datetime_id', how='left')\
                       .merge(pickup_location, on='pickup_location_id', how='left')\
                       .merge(dropoff_location, on='dropoff_location_id', how='left')\
                       .merge(passenger_count, on='passenger_count_id', how='left')\
                       .merge(payment_type, on='payment_type_id', how='left')\
                       .merge(rate_code, on='rate_code_id', how='left')\
                       .merge(trip_distance, on='trip_distance_id', how='left')\
                       .merge(data, on='trip_id', how='left')\
                        [['VendorID', 'datetime_id', 'pickup_location_id', 'dropoff_location_id','rate_code_id',
                          'payment_type_id','passenger_count_id','trip_distance_id', 'store_and_fwd_flag',
                          'payment_type_id', 'fare_amount', 'extra', 'mta_tax', 'tip_amount', 'tolls_amount',
                           'improvement_surcharge', 'total_amount']]
```
